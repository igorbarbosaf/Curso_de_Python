```python
while True:

    print("\nCalculadora Simples")
    print("1 - Adição")
    print("2 - Subtração")
    print("3 - Multiplicação")
    print("4 - Divisão")
    print("5 - Sair")

    opcao = input("Escolha um número (1-5): ")

    if opcao == "5":

        print("Até mais!")

        break

    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))

    if opcao == "1":

        restult = num1 + num2

        print("Resultando é", restult)
    
    elif opcao == "2":

        restult = num1 - num2

        print("Resultando é", restult)
    
    elif opcao == "3":

        restult = num1 * num2

        print("Resultando é", restult)

    elif opcao == "4":

        if num2 != 0:

            restult = num1 / num2

            print(f"Resultando é {restult:.2f}")

        else:

            print("Erro! Divisão por zero")
    else:

        print("Opção Inválida! Tente novamente.")
```
