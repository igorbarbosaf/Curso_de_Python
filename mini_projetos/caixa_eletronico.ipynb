```python
saldo = 1000

while True:

    print("\nCaixa Eletrônico")
    print("1 - Ver Saldo")
    print("2 - Depositar dinheiro")
    print("3 - Sacar dinheiro")
    print("4 - Sair")

    opcao = input("Escolha uma das opções (1-4): ")

    if opcao == "1":

        print(f"Atualmente você tem: R$ {saldo:.2f}")

    elif opcao == "2":

        deposito = float(input("Digite o valor a ser depositando: "))

        if deposito > 0:

            saldo += deposito

            print(f"Depósito do valor R$ {deposito:.2f} realizando com sucesso.")

        else:

            print("Depósito inválido!")

    elif opcao == "3":

        saque = float(input("Digite o valor a ser sacado: R$ "))

        if saque > 0 and saque <= saldo:

            saldo -= saque

            print(f"Saque do valor R$ {saque} realizado com sucesso.")

        else:

            print("Valor inválido ou saldo insuficiente!")

    elif opcao == "4":

        print("Obrigado por ultilizar nosso caixa eletrônico. Até mas...")

        break

    else:

        print("Opção inválida! Porfavor escolha entre (1-4)")
```
