```python
#Pratica utilizado funções 1 - Fábrica de Mensagens

# Criando uma função main com pâramentro (tipo)
def fabrica_de_mensagens(tipo):

    # Função de erro
    def msg_erro():
        return "Ocorreu um erro!"

    # Função de sucesso
    def msg_sucesso():
        return "Operação concluida com sucesso."

    # Função de aviso
    def msg_aviso():
        return "Aviso! operação sem muita informações"

    def msg_desconhecido():
        return "Tipo não reconhecido"
    
    # Usando a condição para puxar as funções
    if tipo == "erro":
        return msg_erro

    elif tipo == "sucesso":
        return msg_sucesso
    
    elif tipo == "aviso":
        return msg_aviso

    else:
        return msg_desconhecido

# Testes
msg_erro = fabrica_de_mensagens("erro")
print(msg_erro()) #Ocorreu um erro!

msg_sucesso = fabrica_de_mensagens("sucesso")
print(msg_sucesso()) #Operação concluida com sucesso

msg_aviso = fabrica_de_mensagens("aviso")
print(msg_aviso()) #Aviso! operação sem muita informações.

msg_desconhecido = fabrica_de_mensagens("qualquercoisa")
print(msg_desconhecido())
```

```python
#Pratica utilizado funções 2 - Calculadora Avançada com args

# Função principal que cria e retorna funções dependendo da operação escolhida
def calculadora_avancada(operacao):

    # Começa com o primeiro número
    def soma(*numeros):
        # sum() já soma tudo automaticamente
        return sum(numeros)
    
    # Função para SUBTRAÇÃO com *args
    def subtracao(*numeros):
        resultado = numeros[0]

        # Subtrai todos os outros
        for num in numeros[1:]:
            resultado -= num

        return resultado
    # Função para MULTIPLICAÇÃO com *args
    def multiplicacao(*numeros):
        resultado = 1 # Multiplicação começa com 1

        for num in numeros:

            resultado *= num
        return resultado

    # Função de DIVISÃO com proteção
    def divisao(*numeros):
        resultado = numeros[0]

        # Divide pelo restante
        for num in numeros[1:]:

            if num == 0:

                # Proteção para divisão por zero
                raise ValueError("Erro: divisão por zero.")

            resultado /= num
        return resultado

    # Função de POTÊNCIA com múltiplos números
    # Ex: potencia(2, 3, 2) = ((2**3)**2)
    def potencia(*numeros):
        resultado = numeros[0]

        for num in numeros[1:]:
            resultado **= num

        return resultado

     # Retornando a função correta baseada na operação pedida
    if operacao == "soma":

        return soma
    
    elif operacao == "subtracao": # sem acento para facilitar

        return subtracao

    elif operacao == "multiplicacao":

        return multiplicacao

    elif operacao == "divisao":

        return divisao

    elif operacao == "potencia":

        return potencia
    else:
        # Caso a operação não exista
        return "Operação inválida!"

# Testes - Usando as funções criadas

# Primeiro pegamos a função desejada
soma = calculadora_avancada("soma")
# Depois chamamos ela com os números
print(soma(1, 2, 3, 4))  # 10

subtrair = calculadora_avancada("subtracao")
print(subtrair(10, 2, 1))  # 7

multiplicar = calculadora_avancada("multiplicacao")
print(multiplicar(2, 3, 4))  # 24

dividir = calculadora_avancada("divisao")
print(dividir(20, 2, 2))  # 5

pot = calculadora_avancada("potencia")
print(pot(2, 3))  # 8
```
